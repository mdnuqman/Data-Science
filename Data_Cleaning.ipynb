```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/realpython/python-data-cleaning/refs/heads/master/Datasets/BL-Flickr-Images-Book.csv")
data.head()
```

```python
to_drop = ["Edition Statement",
           "Corporate Author",
           "Corporate Contributors",
           "Former owner",
           "Engraver",
           "Contributors",
           "Issuance type",
           "Shelfmarks"]
#data.drop(to_drop, inplace=True, axis=1)
data.drop(columns=to_drop, inplace=True)
```

```python
data.head()
```

```python
data["Identifier"].is_unique
```

```python
if 'Identifier' in data.columns:
    data.set_index("Identifier", inplace=True)
data.head()
```

```python
data.loc[1905:, "Date of Publication"].head(10)
```

```python
regex = r'^(\d{4})'
```

```python
extr = data["Date of Publication"].str.extract(regex, expand=False)
extr.head()
```

```python
data["Date of Publication"] = pd.to_numeric(extr)
data["Date of Publication"].dtype
data["Date of Publication"] = data["Date of Publication"].astype('Int64')
```

```python
percentage_of_null = (data["Date of Publication"].isnull().sum()/len(data))*100
print(f"{percentage_of_null:.2f}%")
```

```python
pub = data["Place of Publication"]
london = pub.str.contains("London")
london[:5]
```

```python
oxford = pub.str.contains("Oxford")
```

```python
data["Place of Publication"] = np.where(london, "London",
                                        np.where(oxford, "Oxford",
                                                 pub.str.replace("-"," ")))
```
